```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt 
from sklearn.preprocessing import normalize

%matplotlib inline
# Loading data in
data = sio.loadmat('./train.mat')
images =data['train_images']
labels = data['train_labels']

# Reorganize and preprocess the data, now every column is a data point
X = np.swapaxes(images, 0, 2)
X = np.reshape(X, (60000, 784)).T
y = np.swapaxes(labels, 0, 1)[0]
X = normalize(X * 1.0)
```

```python
print(X.shape, y.shape)
plt.imshow(np.reshape(X[:, 4], (28, 28)))
```

```python
# Now define the most primitive functions
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

def sigmoid_gradient(x):
    return sigmoid(x) * (1 - sigmoid(x))

def tanh(x):
    return np.tanh(x)

def tanh_gradient(x):
    return 1 - tanh(x) * tanh(x)

def relu(x):
    return [i if i > 0 else 0 for i in x]

def relu_gradient(x):
    return [1 if i > 0 else 0 for i in x]
```

```python
x = np.array([-1, 1, 2])
print(relu(x), relu_gradient(x))
```

```python
# Now define the two cost functions
def forward_params(x, W_1, W_2):
    # This is the cost function for a single point
    # J = \sum_{k=1}^K (y_k - h_k(x))
    # x is an array, of size (784, )
    # W_1, W_2 are weight matrices for the first, second layer respectively
    a_0 = np.append([np.ones(len(x[0]))], x, axis=0)
    z_1 = np.dot(W_1, a_0)
    a_1 = np.append([np.ones(len(z_1[0]))], tanh(z_1), axis=0)
    z_2 = np.dot(W_2, a_1)
    a_2 = sigmoid(z_2)
    return (a_0, z_1, a_1, z_2, a_2)

# TODO: Change it to applying gradient implicitly
def gradient_squared(y, W_2, a_0, z_1, a_1, z_2, a_2):
    # Create a vector of length 10 where for y
    label = y
    
    # Make delta_2 a column vector
    Delta_1 = np.zeros((len(a_1)-1, len(a_0)))
    Delta_2 = np.zeros(W_2.shape)
    for i in range(0, len(a_2[0])):
        delta_2 = -(label[:, i] - a_2[:, i]) * sigmoid_gradient(z_2[:, i])
        delta_1 = np.dot(W_2.T[1:], delta_2) * tanh_gradient(z_1[:, i])
        Delta_2 += np.outer(delta_2, a_1[:, i])
        Delta_1 += np.outer(delta_1, a_0[:, i])
    print("a_1.shape = " + str(a_1.shape)+" delta_2.shape = " + str(delta_2.shape))
    return (Delta_1, Delta_2)

def random_initialize(nin, nout):
    # This function randomly initialize a matrix of size nin * nout
    eps = 0.12
    W = np.random.rand(nin, nout) * 2 * eps - eps
    return W
```

```python
W1 = random_initialize(500, 784+1)
W2 = random_initialize(784, 500+1)
(a_0, z_1, a_1, z_2, a_2) = forward_params(X[:, 0:2500], W1, W2)
(d1, d2) = gradient_squared(X[:, 0:2500], W2, a_0, z_1, a_1, z_2, a_2)
```
