```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt 
from sklearn.preprocessing import normalize

%matplotlib inline
# Loading data in
data = sio.loadmat('./train.mat')
images =data['train_images']
labels = data['train_labels']

# Reorganize and preprocess the data, now every column is a data point
X = np.swapaxes(images, 0, 2)
X = np.reshape(X, (60000, 784)).T
y = np.swapaxes(labels, 0, 1)[0]
X = normalize(X * 1.0)
```

```python
# Now define the most primitive functions
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

def sigmoid_gradient(x):
    return sigmoid(x) * (1 - sigmoid(x))

def tanh(x):
    return np.tanh(x)

def tanh_gradient(x):
    return 1 - tanh(x) * tanh(x)

def relu(x):
    return [i if i > 0 else 0 for i in x]

def relu_gradient(x):
    return [1 if i > 0 else 0 for i in x]
```

```python
# Now define the two cost functions
def forward_params(x, W_1, W_2):
    # This is the cost function for a single point
    # J = \sum_{k=1}^K (y_k - h_k(x))
    # x is an array, of size (784, )
    # W_1, W_2 are weight matrices for the first, second layer respectively
    a_0 = np.append([np.ones(len(x[0]))], x, axis=0)
    z_1 = np.dot(W_1, a_0)
    a_1 = np.append([np.ones(len(z_1[0]))], tanh(z_1), axis=0)
    z_2 = np.dot(W_2, a_1)
    a_2 = sigmoid(z_2)
    return (a_0, z_1, a_1, z_2, a_2)

# TODO: Change it to applying gradient implicitly
def gradient_squared(y, W_2, a_0, z_1, a_1, z_2, a_2):
    # Create a vector of length 10 where for y
    label = y
    
    # Make delta_2 a column vector
    Delta_1 = np.zeros((len(a_1)-1, len(a_0)))
    Delta_2 = np.zeros(W_2.shape)
    for i in range(0, len(a_2[0])):
        delta_2 = -(label[:, i] - a_2[:, i]) * sigmoid_gradient(z_2[:, i])
        delta_1 = np.dot(W_2.T[1:], delta_2) * tanh_gradient(z_1[:, i])
        Delta_2 += np.outer(delta_2, a_1[:, i])
        Delta_1 += np.outer(delta_1, a_0[:, i])
    print("a_1.shape = " + str(a_1.shape)+" delta_2.shape = " + str(delta_2.shape))
    return (Delta_1, Delta_2)

def random_initialize(nin, nout):
    # This function randomly initialize a matrix of size nin * nout
    eps = 0.12
    W = np.random.rand(nin, nout) * 2 * eps - eps
    return W
```

```python
W1 = random_initialize(500, 784+1)
W2 = random_initialize(784, 500+1)
(a_0, z_1, a_1, z_2, a_2) = forward_params(X[:, 0:2500], W1, W2)
(d1, d2) = gradient_squared(X[:, 0:2500], W2, a_0, z_1, a_1, z_2, a_2)
```

```python
# Applying the thresholding, or doing backward solve directly
def tri_solve(A, B, lamb, b, back=True):
    # A is upper triangular, 
    # B is upper trapozoidal, 
    # lamb is a threshold (\sqrt(\lambda))
    # This does a backward solve for A\kron B x = b
    # Also returns the positions of inserted zero rows
    n = A.shape[0]
    (k, l) = B.shape
    x = np.zeros((n*l, 1))
    
    prod = lin.kron(A, B)
    if back:
        for i in range(n):
            prod=np.insert(prod, i*l+k, np.zeros((l-k, n*l)), axis=0)
    else:
        for i in range(n):
            prod=np.insert(prod, i*l+k, np.zeros((n*l, l-k)), axis=1)
    
    for i in range(n*l):
        prod[i,i] = max(prod[i, i], lamb)
    
    print prod
    if back:
        return lin.solve_triangular(prod, b)
    else:
        return lin.solve_triangular(prod, b, lower=True)


# Need testing
def kron_mult(k, l, n, N, M, y, back=True):
    # k, l, n indicates the zeros blocks we have to insert/delete from y
    # This function calculates N\kron My, with zeros blocks handled
    if back == True:
        y.reshape((M.shape[0], N.shape[0]), order='F')
        y = np.dot(np.dot(M.T, y), N).flatten('F')
    else:
        # Deleting zeros first
        for i in range(n):
            y = np.delete(y, range(i*l+k, (i+1)*l), axis=0)
        y.reshape((M.shape[1], N.shape[1]),order='F')
        y = np.dot(np.dot(M, y), N.T).flatten('F')
        
    for i in range(n):
        y = np.insert(y, i*l+k, np.zeros((l-k, 1)), axis=0)
    return y                  

# Need testing, solve the large triangular system
def solve_R(L, MS, NS, lamb, b, back=True):
    # L = number of layers
    # MS = collection of the M matrices
    # NS = collection of the N matrices
    # lamb = threshold
    # b = right hand side, in a blocked dict way
    # REMARK: The block ordering or x and b is consistent with the inner and outer loop
    x = {}
    if back == True:
        for r in range(0, L):
            for c in range(0, r):
                M = MS[(r, c)]
                N = NS[(r, c)]
                n = NS[(c, c)].shape[0]
                (k, l) = MS[(c, c)].shape
                b[r] = b[r] - kron_mult(k, l, n, N, M, x[c])
            M = MS[(r, r)]
            N = NS[(r, r)]
            x[r] = tri_solve(N, M, lamb, b[r])
    else:
        for c in range(L-1, -1, -1):
            for r in range(L-1, c, -1):
                M = MS[(r, c)]
                N = NS[(r, c)]
                n = NS[(c, c)].shape[0]
                (k, l) = MS[(c, c)].shape
                b[c] = b[c] - kron_mult(k, l, n, N, M, x[r], False)
            M = MS[(c, c)]
            N = NS[(c, c)]
            x[c] = tri_solve(N, M, lamb, b[c], False)
    return x
```
